```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import re
```

```python
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
res = requests.get(url)
res.encoding = res.apparent_encoding
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
main_text = soup.find("div", class_="main_text").get_text()
```

```python
text = re.sub(r"\s+", "", main_text)
```

```python
text = re.sub(r"《.*?》", "", text)
text = re.sub(r"［.*?］", "", text)
```

```python
print(text[:200])
```

```python
first_sentence = re.split("。", text)[0]
print(first_sentence)
```
